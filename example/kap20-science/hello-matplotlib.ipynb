```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.plot(x, y1, '-', x, y2, '--')
```

```python
f  = lambda x: np.sin(x) * np.exp(-0.1 * x)
x  = np.linspace(0, 25, 1000)
y  = f(x)
plt.plot(x, y)
```

```python
f  = lambda x: math.sin(x) * math.exp(-0.1 * x)
vf = np.vectorize(f)
x  = np.linspace(0, 25, 1000)
y  = vf(x)
plt.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x', ylabel='f(x)', title='Gedämpfte Schwingung')
ax.grid()
```
